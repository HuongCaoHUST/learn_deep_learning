```python
import torch
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Net (nn.Module):
    def __init__ (self):
        super (Net, self).__init__()
        self.conv1 = nn.Conv2d(1,6,3) #(input_ch, output_ch, stride)
        self.conv2 = nn.Conv2d(6,16,3)
        self.fc1 = nn.Linear(16*6*6,120)
        self.fc2 = nn.Linear(120,84)
        self.fc3 = nn.Linear(84,10)

    def forward (self,x):
        x = self.conv1(x)
        x = F.relu(x)
        x = F.max_pool2d(x,2)
        x = self.conv2(x)
        x = F.relu(x)
        x = F.max_pool2d(x,2)
        x = x.view(-1,self.num_flat_features(x))

        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        return x

    def num_flat_features(self,x):
        size = x.size()[1:]
        num_features = 1
        for s in size:
            num_features *= s
        return num_features
```

```python
net=Net()
print(net)
```
